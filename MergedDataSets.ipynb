```python
weather_df = pd.read_csv("NY_weather.csv")
weather_df.head()
```

```python
traffic_df = pd.read_csv("traffic_counts.csv")
traffic_df.head()
```

```python
# Create 'datetime' column in traffic_df
traffic_df['datetime'] = pd.to_datetime(traffic_df[['Yr', 'M', 'D', 'HH', 'MM']].astype(str).apply(lambda x: ''.join(x), axis=1), format='%Y%m%d%H%M', errors='coerce')

# Check for NaT values
print(traffic_df[traffic_df['datetime'].isna()])

# Handle NaT values (e.g., drop rows or impute)
traffic_df = traffic_df.dropna(subset=['datetime'])
```

```python
weather_df['datetime'] = pd.to_datetime(weather_df['datetime'], errors='coerce')
```

```python
weather_df = weather_df.set_index('datetime')
traffic_df = traffic_df.set_index('datetime')
```

```python
merged_df = pd.merge(weather_df, traffic_df, left_index=True, right_index=True, how='inner')
```

```python
print(merged_df.head())
print(merged_df.info())
```

```python
merged_df = merged_df.drop(columns=['SegmentID', 'RequestID', 'sealevelpressure', 'street', 'fromSt', 'toSt', 'Direction', 'stations', 'icon', 'seversisk', 'winddir', 'dew', 'solarradiation', 'solarenergy', 'visibility'], errors='ignore')
```

```python
print(merged_df.head())
print(merged_df.info())
```

```python
merged_df = merged_df.drop(columns=['severerisk', 'cloudcover', 'windgust', 'windspeed', 'precipprob'], errors='ignore')
```

```python
print(merged_df.head())
print(merged_df.info())
```
